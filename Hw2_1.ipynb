Прочитайте дані за допомогою методу `read_html` з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)" [за посиланням](https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8 "Коефіцієнт народжуваності в регіонах України (1950—2019")

```python
#presetings

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

URL = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8"
tables = pd.read_html(URL, match="Коефіцієнт народжуваності в регіонах України")
df = tables[0]
df
```

2.1.1 Вивести перші рядки таблиці за допомогою методу `head`

```python
print(df.head())
```

2.1.2 Визначте кількість рядків та стовпців у датафреймі (атрибут `shape`)

```python
print(df.shape)
```

2.1.3 Замініть у таблиці значення "—" на значення NaN

```python
new_data = df.replace(to_replace="—", value=np.nan)
new_data
```

2.1.4 Визначте типи всіх стовпців за допомогою `dataframe.dtypes`

```python
df.dtypes
```

```python
new_data = new_data.astype(
    {
        "1950": "float64",
        "1960": "float64",
        "1970": "float64",
        "1990": "float64",
        "2000": "float64",
        "2012": "float64",
        "2014": "float64",
        "2019": "float64",
    }
)
new_data
```

2.1.5 Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
update_data = new_data.fillna(0, inplace=False)
update_data
```

2.1.6 Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи `isnull` та `sum`)

```python
missing_prc = new_data.apply(lambda column: column.isnull().sum() / len(new_data) * 100)
round(missing_prc, 2)
```

2.1.7 Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
update_data = update_data.drop([27])
update_data
```

2.1.8 Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод `fillna`)

```python
for column in new_data.columns[1:]:

  mean_value = new_data[column][1:].mean()
  new_data[column].fillna(mean_value, inplace=True)

new_data.round(1)
```

2.1.9 Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
top_birth_rate_regions = new_data[new_data['2019'] > new_data['2019'].mean()]['Регіон']
print(top_birth_rate_regions.drop([27]))
```

2.1.10 У якому регіоні була найвища народжуваність у 2014 році?

```python
max_birth_rate_region_2014 = new_data.loc[new_data['2014'].idxmax(), 'Регіон']
print(max_birth_rate_region_2014)
```

2.1.11 Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
new_data.plot(x='Регіон', y='2019', kind='bar', legend=False)
plt.title('Народжуваність по регіонах у 2019 році')
plt.xlabel('Регіон')
plt.ylabel('Народжуваність')
plt.show()
```
