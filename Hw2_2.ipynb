2.4 Порахуйте, яка частка пропусків міститься в кожній колонці

```python
missing_prc = df.apply(lambda column: column.isnull().sum() / len(df) * 100)
missing_prc
```

2.5 Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"

```python

lang_column = df['Язык.программирования']
columns = df.columns[df.isnull().sum() == 0]

columns = columns.append(pd.Index(['Язык.программирования']))
new_df = df[columns]

new_df['Язык.программирования'] = lang_column
new_df
```

2.6 Знову порахуйте, яка частка пропусків міститься в кожній колонці

```python
missing_prc = new_df.apply(lambda column: column.isnull().sum() / len(new_df) * 100)
missing_prc
```

2.7 Видаліть усі рядки у вихідній таблиці за допомогою методу dropna

```python
new_df = new_df.dropna()
new_df
```

2.8 Визначте новий розмір таблиці за допомогою методу `shape`

```python
new_df.shape
```

2.9 Створіть нову таблицю `python_data`, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування `Python`

```python
python_data = new_df[new_df['Язык.программирования'] == 'Python']
python_data
```

2.10 Визначте розмір таблиці `python_data` за допомогою методу `shape`

```python
python_data.shape
```

2.11 Використовуючи метод `groupby`, виконайте групування за стовпчиком "Посада"


```python
python_data = python_data.groupby('Должность')
python_data
```

2.12 Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу `agg` і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
aggregated_data = python_data.agg({'Зарплата.в.месяц': ['min', 'max']})
aggregated_data
```

2.13 Створіть функцію `fill_avg_salary`, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу `apply` та створіть новий стовпчик "avg"

```python
def fill_avg_salary(row):
  return row['Зарплата.в.месяц'].mean()

aggregated_data['avg'] = aggregated_data.apply(fill_avg_salary, axis=1)
aggregated_data
```

2.14 Створіть описову статистику за допомогою методу `describe` для нового стовпчика

```python
aggregated_data['avg'].describe()
```
